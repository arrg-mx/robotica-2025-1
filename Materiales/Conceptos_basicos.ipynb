
## URDF


Los robots son sistemas mecatrónicos de natuzaleza compleja por el tipo de interaccines que realizan con su entorno con el objetivo de realizar una tarea específica. Actualmente, se plantean diferentes enfoques cuyo objetivo es dotar a los manipuladores moviles con la capacidad de manipular y transportar materiales dentro de un entorno estructurado. 
En la mayoría de los casos la manipulación de objetos se realizar utilizando robots seriales, los cuales deben realizar movimientos coordenados en sus artículaciones determinados por algún modelo de manipulación.

El desarrollo e implementación de algún modelo de manipulación con lleva un cierto grado de complejidad, costos y riesgos a ser provados en un robot real, es por ello que dichas pruebas se realizan utilizando modelos de simulaciones computacionales utilizando algún tipo de interfaz.Las interfaces computacionales son herramientas que permiten análizar y/o verificar el comportamiento de un robot que realiza alguna tarea y para ello es necesario generar la información que describe al robot con todos sus elementos y la información que describe a todos los elementos del entorno. 

La información que describe a los robots está contenida en archivos "Unified Robot Description Format" (URDF), que como su nombre lo indica es un formato unificado utilizado para describir a un robot en formato XML. 

La descripción de los robots seriales utilizando en URDF se compone de diferentes elementos. A continuación, se presentan del desarrollo y la descripcción de un archivo URDF de un eslabón el cual está restringido por una junta de revolución.

### Encabezado y descripción del robot

Los archivos URDF tienen un encabezado el cual indica la versión del formato xml que se está utilizando y de manera opcional el idioma en el cual está codificado, en el ejemplo es "utf-8" (8-bit Unicode Transformation Format). Después del encabezado se encuentra la descripción del robot, el inicio de la descripción del robot debe indicar el nombre del robot, además se debe indicar su limite, como se muestra en el siguiente código:

```xml
<!--Encabezado del archivo-->
<?xml version="1.0" encoding="utf-8"?>

<!--Descripción del robot-->
<robot name="common_link"> <!--Inicio de la descripción del robot-->


</robot> <!--Fin de la descripción el robot-->

```

Dentro de la descripción del robot se colocan los elementos que describen la robot, para construirá un robot compuesto por geometrias primitivas. 

### Definición de eslabón

La parte mecánica de los robots se describen como "links" (eslabones),los cuales describen su visualización, sus propiedades inerciales y sus relaciones de colisión. El código que describe a un eslabón de manera general es:

```xml
    <link name="name">
        <!--Descripción de la parte visual del eslabón-->
        <visual>
            
        </visual>
        <!--Descripción de las relaciones de colision-->
        <collision>
            
        </collision>
        <!--Descripción de la inercial del eslabón-->
        <inertial>
            
        </inertial>
        
    </link>

```

#### base_link

Cualquier descripción de un robot, serial o móvil, comienza con un eslabón base denominado "base_link" que en el caso de los robot seriales es un eslabón fijo el cual no se mueve dentro del espacio de trabajo. A continuación se presenta la definición del base_link conformado por un prisma rectangular 

##### Parte visual

Como su nombre lo indica la parte visual definie la apariencia del eslabón a partir de tres parámetros, la pose del eslabón, su geometría y el su material. La pose del eslabón determina la posición del eslabón con respecto sistema de origen inercial en coordenadas cartesianas, además de su orientación utilizando la convención rpy (yaw, pitch and roll). Para la descripción de los parematros se emplean las unidades del sistema ISO, en el caso de las coordenadas las unidades utilizadas son metros, y para las descripción de la orietación las unidades que se utilizan son radianes. 

```xml
    <origin xyz="traslacion_x traslacion_y traslacion_z" rpy="orient_x orient_y orient_z"/>

```

En el código anterior, describe la posición y orientación de un sistema de referencia asociado con la geometría del eslabón, el cual tiene una:

- traslación sobre el eje x del sistema inerical (traslacion_x),
- una traslación sobre el eje y del sistema inerical (traslacion_y), 
- una traslación sobre el eje z del sistema inerical (traslacion_z),
- una orientación del eje x con respecto al eje x del sistema inercial (orient_x),
- una orientación del eje y con respecto al eje y del sistema inercial (orient_y) y 
- una orientación del eje z con respecto al eje z del sistema inercial (orient_z).


##### Parte geométrica

Está parte define la parte geométrica del eslabón, la cual puede ser definida como una geometría primitiva (cubos, primas, cilindros y esferas) o puede ser definida por una malla exportada de un CAD. Los parámetros de está definición cuenta con multiples opciones, por lo que se recomienda consultar la documentación oficial (http://wiki.ros.org/urdf/Tutorials). 

Para este caso la geometría del robot se define como un prisma donde su arista sobre el eje x mide 0.1 m, su arista sobre el eje y mide 0.12 m y su arista sobre el eje z mide 0.15 m.

```xml
    <geometry>
            <box size="0.1 0.12 0.15"/>             
    </geometry>

```

#####  Definciión del Material

La definición del material consiste determinar el color del eslabón y su transparecia.


```xml
    <material name="color_name">
             <color rgba="red green blue alpha" />
    </material>      
        
```

En el código anterior se define el nombre del color (color_name), la definición del color está dada por la combinación de los colores primarios rojo (red), verde (green), y azul (blue), la intensidad de los colores se introducen con un valor en un rango en 0.0 a 1, la transparencia del eslabón está definida por el parámetro alpha con rango de valores de 0.0 a 1, 0.0 es transparente y 1 es totalmente solido. 

La definición de material puede ser general en la descripción del robot y ser llamada en la definición del eslabón, lo anterior, se presentará en posteriores posteriores ejemplos en los cuales se definen diferentes controladores para mover al eslabón.

#####  Código de la parte visual

El código que define la parte visual es:


```xml
    <link name="name">
        <!--Descripción de la parte visual del eslabón-->
        <visual>
            <origin xyz="traslacion_x traslacion_y traslacion_z" 
                        rpy="orient_x orient_y orient_z"/>
            <geometry>
                <box size="0.1 0.12 0.15"/>             
            </geometry>
            <material name="color_name">
                <color rgba="red green blue alpha" />
            </material> 
        </visual>
      
    </link>

```


##### Parte de colisión

La parte de colisión define los limites físicos del eslabón al interactuar con otros elementos de su entorno. La definición de la colisión de un eslabón se compone de la definición de los parámetros de origen y de los parámetros de definición geometría, las cuales son independientes de los definidos de la parte visual. El código que define la parte de colisión para este ejemplo es:

```xml
    <link name="name">
        <!--Descripción de la parte visual del eslabón-->
        <collision>
            <origin xyz="traslacion_x traslacion_y traslacion_z" 
                        rpy="orient_x orient_y orient_z"/>
            <geometry>
                <box size="x y z"/>             
            </geometry>
        </collision>
      
    </link>

```

##### Parte inercial

La parte inercial, define la inercia lineal del eslabón y la inercial rotacional del eslabón con respecto al centro de masa de este. Por lo anterior, es necesario definir la masa del eslabón, los elementos de la matriz de ineria y la ubicación del centro de masa del eslabón con respecto a su centro geométrico, de igual manera los parametros están dados en sistema ISO, por lo que la masa está en kilogramos (kg) y en kilogramos sobre metros cuadrados ($\frac{kg}{m^2}$). El código que define los elementos inerciales es:

```xml
    <link name="name">
        <!--Descripcion de la inercial del eslabon-->
        <inertial>
            <!--masa del eslabon-->
            <mass value="0.0"/>
            <!--Ubicacion del sistema inercial asociado al centro de
                 masa-->
            <origin xyz="0.0 0.0 0.0" rpy="0.0 0.0 0.0"/>
            <!--Matriz de inercia en el centro de masa-->
            <inertia ixx="0.0" ixy="0.0" ixz="0.0" iyy="0.0" iyz="0.0" 
                     izz="0.0"/> 
        </inertial>
        
    </link>

```

En el caso de la matriz de inercia solo es necesario definir seis elementos ya que en su centro de masa es una matriz simetrica. 


Nota: Para la definición del "base_link" no es necesario definir las propiedades inerciales, en caso de que la definición de este cuente con la definición de inercia el simulador marcara una advertensia al respecto. 


#### Código del eslabón inercial

El Código que define al eslabón inercial es:


```xml
<!--Encabezado del archivo-->
<?xml version="1.0" encoding="utf-8"?>
<!--Descripcion del robot-->
<robot name="common_link"> <!--Inicio de la descripción del robot-->

    <link name="base_link">
        <!--Descripcion de la parte visual del eslabon-->
        <visual>
            <!--Posicion y orientación de origen del eslabon-->
            <origin xyz="0.0 0.0 0.0" rpy="0.0 0.0 0.0"/>
            <!--Definición de la geometria del eslabón-->
            <geometry>
                <box size="0.1 0.12 0.15"/>             
            </geometry>
            <!--Definicion del color del eslabon-->
            <material name="grey">
                <color rgba="0.5 0.5 0.5 0.0" />
            </material>      
        </visual>
        <!--Descripcion de las relaciones de colision-->
        <collision>
            <!--Posicion y orientacion de origen del eslabon-->
            <origin xyz="0.0 0.0 0.0" rpy="0.0 0.0 0.0"/>
            <!--Definicion de la geometria del eslabon-->
            <geometry>
                <box size="0.1 0.12 0.15"/>             
            </geometry>          
        </collision>
        <!--Descripcion de la inercial del eslabon-->
        <inertial>
            <!--masa del eslabon-->
            <mass value="0.5"/>
            <!--Ubicacion del sistema inercial asociado al centro 
                de masa-->
            <origin xyz="0.0 0.0 0.0" rpy="0.0 0.0 0.0"/>
            <!--Matriz de inercia en el centro de masa-->
            <inertia ixx="0.1" ixy="0.0" ixz="0.0" iyy="1.0" 
                     iyz="0.0" izz="1.0"/> 
        </inertial>
    </link>
    
</robot> <!--Fin de la descripción el robot-->

```

Nota: De este código se han eliminado los acentos para eviar errores en su compilación.

### Definiciones de juntas

La configuración mecánica de los robots cuenta con arreglos mecánicos que permite el movimiento relativo entre los eslabones que la integran, estos arreglos se denóminan juntas. Las juntas se clasifican por el tipo de restricción de movimiento que genera entre dos o más eslabones, por lo que las juntas se clasifican en:

- Revolute joint, las juntas de revolución son juntas rotacionales las cuales funjen como las artículaciones de un brazo humano, los parámetros que la definen son el eslabón "hijo" (eslabón que será conectado), eslabón "padre" (eslabón al cual se le conectará el eslabón hijo), el eje de acción de la junta, el límite inferior e superior de movimiento del robot, en radianes, la velocidad máxima de la junta en radianes sobre segundo, el límite de esfuerzo superior del actuador, en Newton por metro, asociado a la junta y la posición y orientación del sistema de referencia asociado a la junta. El código que lo define a esta junta es:

```xml
    
    <joint name="joint_name" type="revolute">
        <parent link="parent_link"/>
        <child link="child_link"/>
        <axis xyz="0 0 0"/>
        <limit lower="-3.14159" upper="3.14159" velocity="50.0" 
               effort="1000.0"/>
        <origin xyz="0 0 0" rpy="0 0 0"/>
    </joint>

```

- Continuos joint, las juntas continuas son juntas rotacionesles cuyo rango de movimiento no tiene un límite, este tipo de juntas son empleadas para definir las llantas de los robots móviles. El código que define a este tipo de juntas es:

```xml
    
    <joint name="joint_name" type="continuous">
        <parent link="parent_link"/>
        <child link="child_link"/>
        <axis xyz="0 0 0"/>
        <origin xyz="0 0 0" rpy="0 0 0"/>
    </joint>

```

- Prismatic joint, las juntas prismáticas las cuales limitan el movimiento de un eslabón para que es solo se pueada mover en la dirección de un eje. Los pámetros que definen a este tipo de juntas son su limite inferior y superior de la junta , en metros, la velocidad máxima de la junta, en metros sobre segundo, y la fuerza generada por el actudor, en Newtons. El código que define a la junta es:

```xml
    
    <joint name="joint_name" type="prismatic">
        <parent link="parent_link"/>
        <child link="child_link"/>
        <axis xyz="0 0 0"/>
        <limit lower="-0.5" upper="0.5" velocity="0.5" effort="1000.0"/>
        <origin xyz="0 0 0" rpy="0 0 0"/>
    </joint>

```

- Fixed joint, las juntas fijas se utilizan para unir dos eslabones y restringir su movimiento. Los parámetros que definen a esta junta son el eslabón padre, el eslabón hijo, la posición y orientación del sistema de referencia asociado a la junta. El código que define a este tipo de juntas es:

```xml
    
    <joint name="joint_name" type="fixed">
        <parent link="parent_link"/>
        <child link="child_link"/>
        <origin xyz="0.0 0.0 0.0" rpy="0.0 0.0 0.0"/>
    </joint>
    
```

## Vizualización en rviz

RVIZ es una herramienta la cual permite interactuar con un robot mediante el flujo de información relacionada con el control de sus motores, la información que describe el estado del robot y la información del entorno. ### Plugin

Un "plugin"  complementa la descripción de robot dentro del simulador de Gazebo, al asociar a un o a un conjunto de sistemas de referencia esquemas de control según la configuración del robot o definiendo algún tipo de sensor. La definición de cada plugin varia conforme a la configuración mecánica de cada robot y al tipo de sensores que tenga el robot. 

En el caso de los robots seriales, la definición de los plugins definen los sensores y de las características de los actuadores al igual que los esquemas de control de estos, esto último es lo que dificulta el simular a los robots seriales. 

Para este caso, se utilizaran los plugs de ROS 2 para definir a la junta del eslabón con un control de trayectoria en el espacio de las juntas del robot.

Los plugin´s que definen a las juntas del robot son:

- joint_state_publisher_controller, que se encarga de publicar en el controlador de las juntas.

- joint_pose_trayectory_controller, define el tipo de controlador de la junta.

El código correspondiente es:

```xml

    <gazebo>
        <plugin name="joint_state_publisher_controller"
                filename="libgazebo_ros_joint_state_publisher.so">
            <!-- Frecuencia de actualizacion del plugin Hertz -->
            <update_rate>10</update_rate>
            <!-- Lista de las juntas que funcionaran con el controlador 
                    -->
            <joint_name>base_link_to_link_1</joint_name>
        </plugin>
     </gazebo>
     <!--Definicion del controlador-->
    <gazebo>
        <plugin name="joint_pose_trajectory_controller"
                filename="libgazebo_ros_joint_pose_trajectory.so">
            <!-- Frecuencia de actualizacion del controlador Hertz -->
            <update_rate>2</update_rate>
        </plugin>
    </gazebo>
``` 

En el código anterior se define un control *joint_pose_trajectory_controller* el cual genera de manera autómatica el Nodo */joint_pose_trajectory_controller* el cual genera el Tópico */joint_state_publisher_controller* sobre el cual se *publica* la posición en la cual se requiere mover la junta en radianes. El formato del tópico es el siguiente:


```text
$ ros2 topic pub -1 /set_joint_trajectory trajectory_msgs/msg/JointTrajectory '{header: {frame_id: world}, joint_names:[base_link_to_link_1],points:[{positions:{-1.57}} ]}'
```
La entrada anterior, manda la eslabón para posicionarse $\frac{\pi}{2}$($-1.57$) $rad$

# Configuración básica de una junta 

El controlar un robot tiene como objetivo el coordinar el movimiento de sus actuadores con el fin de que este sea capaz de realizar una determinada tarea. En el caso de los robots seriales, la coordinación de los actuadores de sus juntas están sujetas a la definición de su controlador, los controladores que se pueden implementar en las juntas de un robot son:

- position_controller (controladores de posición de posición).

- velocity_controller (controladores de velocidad).

- joint_trajectory_controller (controladores de esfuerzos)-


Para utilizar los controladores anteriores es necesario constar de los siguientes elementos:

1. La interfase de control, la cual le el estado de las juntas y define cual o cuales son los comandos de control de la junta.
2. Los parámetros del controlador, son las definiciones partículares que definen a cada junta como el nombre, el tipo de junta, los límites de su funcionamiento.
3. El control manager, activa o desactiva el funcionamiento de la junta, o detiene su acción.


Para la implementación de los elementos anteriores es necesario realizar su implementación de la siguiente manera (la implementación es para simulaciones en Gazebo Classic, https://control.ros.org/rolling/doc/gazebo_ros2_control/doc/index.html):

1. Actuzalizar ubuntu

```txt
$ sudo apt-get update && sudo apt upgrade
```

2. Instalar *control*

```txt
$ sudo apt-get install ros-humble-gazebo-ros2-control
```

3. Instalar los controladores

Controlador de posición, velocidad y esfuerzo
```txt
$ sudo apt-get install ros-humble-position-controllers ros-humble-velocity-controllers ros-humble-joint-trajectory-controller 
```

4. Instalar ROS 2 Control
```txt
$ sudo apt-get install ros-humble-ros2-control
```

5. Instalar el Control Manager
```txt
$ sudo apt-get install ros-humble-controller-manager
```

Con la instalación anterior es posible implementar en la simulaciones en Gazebo controladores de posición, controladores de velocidad y controladores de trayectorias sobre las juntas, los cuales deben ser implementados desde la configuración del URDF del robot. La implementación básica de controlador es la siguiente:


```xml

<ros2_control name="GazeboSystem" type="system">
    <hardware>
        <plugin>gazebo_ros2_control/GazeboSystem</plugin>
    </hardware>
    <joint name="slider_to_cart">
        <command_interface name="effort">
            <param name="min">-1000</param>
            <param name="max">1000</param>
        </command_interface>
        <state_interface name="position">
            <param name="initial_value">1.0</param>
        </state_interface>
        <state_interface name="velocity"/>
        <state_interface name="effort"/>
    </joint>
</ros2_control>

```

El código anterior define un controlador en el sistema de Gazebo del sistema, y carga el plugin del controlador de ROS 2. En el caso de la definición de la junta tiene los siguientes parámetros:

- Nombre de la junta "slider_to_cart"
- Interface de comando, en este caso es del tipo "effort" (esfuerzo) por lo que la entrada para controlar la junta es una fuerza.
- Parámetros de la junta, valor mínimo de entrada -1000 N y valor máximo de entrada de junta 1000 N.
- Estado inicial de la junta, en este caso es de 1.0 m.
- Las interface de estado de la junta, en este caso la configuración del estado de la junta permite conocer la posición, la velocidad y el esfuerzo presentes sobre la junta.


## Control de posición de una junta rotacional

Para controlar la junta del eslabón mediante un controlador de posición es necesario agregar su configuración dentro de la interface de Gazebo al agregar el siguiente código:

```xml

<ros2_control name="GazeboSystem" type="system">
        <hardware>
            <plugin>gazebo_ros2_control/GazeboSystem</plugin>
        </hardware>
        <joint name="base_link_to_link_1">
            <command_interface name="position"/>
            <state_interface name="position"/>
            <state_interface name="velocity"/>
            <state_interface name="effort"/>
        </joint>
</ros2_control>

```

Donde:

```xml

<ros2_control name="GazeboSystem" type="system">

</ros2_control>

```
Contiene la descripción del controlador de las juntad dentro de gazebo.

```xml
        <hardware>
            <plugin>gazebo_ros2_control/GazeboSystem</plugin>
        </hardware>
```
Carga el pluging de control dentro de Gazebo.

```xml
        <joint name="base_link_to_link_1">
            <command_interface name="position"/>
            <state_interface name="position"/>
            <state_interface name="velocity"/>
            <state_interface name="effort"/>
        </joint>
```

En esta parte se define las interfaces de una junta mediante *name="base_link_to_link_1"*, *command_interface* indica mediante tipo de entrada se controlará la junta para este caso es  por posición (*position*), *state_interface* publica el estados cinemáticosde posición y velocidad de junta (*position* y *velocity*) y el par (*effort*) de la junta. En el caso de los robots seriales se deben definir todas las juntas dentro de esta sección.



Ya con las juntas definidas es necasario cargar los parametros del controlador los cuales deben ser específicados mediante el siguiente código.

```xml

    <gazebo>
        <plugin filename="libgazebo_ros2_control.so" name="gazebo_ros2_control" >
            <parameters>$(find ${package_name})/config/basic_link_position_controller.yaml</parameters>
        </plugin>
    </gazebo>

```

El código anterior carga y contruye el controladodor del robot a partir de la libreria *libgazebo_ros2_control.so* mediate los parámetros específicados en el archivo *basic_link_position_controller.yaml* el cual debe estar dentro de un metapaquete *config*. Por lo anterior es necesario agregar al paquete *basic_link_description* el metá paquete y el archivo.

```text

 basic_link_description/
    ├── config/
    │   └── basic_link_position_controller.yaml
    
```

Los parámetros que definen al controlador de posición implementado son los siguientes:

```xml
controller_manager:
  ros__parameters:
    update_rate: 100  # Hz

    basic_link_position_controller:
      type: forward_command_controller/ForwardCommandController

    joint_state_broadcaster:
      type: joint_state_broadcaster/JointStateBroadcaster

basic_link_position_controller:
  ros__parameters:
    joints:
      - base_link_to_link_1
    interface_name: position
```

El código anterior le indica al controller manager los parámetros del controlador, en este caso son:

- Que el controlador trabaje con una frecuencia de 100 Hz
- El nombre que tendrá el controlador y el tipo, en este casi el nombre es *basic_link_position_controller* y el tipo es *forward_command_controller/ForwardCommandController*
- Genera y confgura el broadcaster para publicar el estado de las juntas.
- Después carga los parámetros y configura el controlador sobre las juntas enlistadas y se específica la interface de control que en este caso es *position*. Para este caso no está configurada la retroalimentación de velocidad y de esfuero de la junta por lo que al ejecutarse el controlador no se tendran esas lecturas.
